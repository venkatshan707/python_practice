### what is the use of if \_\_name__ == "\_\_main__"?

if \_\_name__ == "\_\_main__": is commonly referred to as an "execution guard" or "main guard" in Python.

>It's a conditional statement that checks if the current script is being run as the main program.

>This condition is used to control the execution of code within a Python script. It ensures that **certain parts of the code are only executed when the script is run directly as the main program**, but not when it's imported as a module into another script.

Example :
when we import a module kansas: 

in the kansas.py, if we simply called fucntion  randomfunfact():

 with out mentioning it under the condition 'if \_\_name\_\_ == "\_\_main\_\_":' 
 
 then randomfunfact() is called.

But actually it should not run the function when we import. To avoid this we are mentioning it under 
'if \_\_name\_\_ == "\_\_main\_\_":'.

 **So that it will run only if we call the function not importing it's package(like kansas.randomfunfact())**

Here's why it's used and its significance:

**Execution Control:** It allows you to specify which parts of the script should be executed when the script is run directly from the command line or as a standalone program. This helps in organizing your code and separating the main program logic from other definitions or statements that may be used when the script is imported as a module.

**Prevent Side Effects:** Without the if __name__ == "__main__": guard, any code outside functions or classes in the script would be executed as soon as the script is imported as a module. This can lead to unintended side effects or execution of code that you might not want to run when the script is imported.
